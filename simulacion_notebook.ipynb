# Preámbulo


```python
# Cargar librerías y fijar una semilla para asegurar replicabilidad
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(505)
```

# Funciones

```python
def rw(steps):
    """Produce una marcha aleatoria según 
    el número de pasos que se pasa como argumento. Se 
    retorna una lista con los pasos, el camino, y la
    posición final (fin del camino).

    Args:
        steps (int): N, el número de pasos.

    Returns:
        list: Lista con 
        1. La distancia a desplazarse en cada paso
        2. El camino
        3. Posición final
    """
    pasos = [np.random.choice([-1,1]) for n in range(steps)]
    camino = np.concatenate([[0],pasos]).cumsum(0)
    return [pasos, camino ,camino[-1:][0]]
```

# Implementación

```python
# Parámetros y simulación
N = 10000
camino = rw(N)[1]

# Producir gráfica
fig = plt.figure(figsize=(8,4),dpi=200)
ax = fig.add_subplot(111)
ax.scatter(np.arange(N+1), camino, c='blue',alpha=0.25,s=0.05)
ax.plot(camino,c='blue',alpha=0.5,lw=0.5,ls='-')
ax.plot(0, 0, c='red', marker='+')
ax.plot(N, camino[-1:], c='black', marker='o')
plt.title("Marcha aleatoria en 1D")
plt.ylabel("Posicion")
plt.xlabel("Tiempo")
```
